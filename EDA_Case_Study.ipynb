```python
# Examine the distribution of unique cataegories in categorical columns
categorical_cols = df[["BRId", "GenderId", "IAId", "Amount_of_Credit_Cards", "Nationality", "Occupation", "Fee_Structure", "Loyalty_Classification", "Properties_Owned", "Risk_Weighting", "Income_Band"]].columns

for col in categorical_cols:
  print(f"Value Counts for '{col}':")
  display(df[col].value_counts())
```

# Univariate Analysis

```python
for i, predictor in enumerate(df[["BRId", "GenderId", "IAId", "Amount_of_Credit_Cards", "Nationality", "Occupation", "Fee_Structure", "Loyalty_Classification", "Properties_Owned", "Risk_Weighting", "Income_Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor)
```

```python
# HIstplot of value counts for different Occupation

for col in categorical_cols:
  if col == "Occupation":
    continue
  plt.figure(figsize=(8,4))
  sns.histplot(df[col])
  plt.title('Histogram of Occupation Count')
  plt.xlabel(col)
  plt.ylabel("Count")
  plt.show()
```

# Numerical Analysis

```python
numerical_cols = ['Estimated_Income', 'Superannuation_Savings', 'Credit_Card_Balance', 'Bank_Loans', 'Bank_Deposits', 'Checking_Accounts', 'Saving_Accounts', 'Foreign_Currency_Account', 'Business_Lending']

sns.set_style("whitegrid")
sns.set_palette("Set2")
# Univariate analysis and visualization
plt.figure(figsize=(15,10))
for i,col in enumerate(numerical_cols):
  plt.subplot(4,3,i+1)
  sns.histplot(df[col],kde=True)
  plt.title(col)
plt.show()


```

# HeatMap

```python
sns.set(style="white", font_scale=1.2)

numerical_cols = [
    'Estimated_Income', 'Superannuation_Savings', 'Credit_Card_Balance',
    'Bank_Loans', 'Bank_Deposits', 'Checking_Accounts',
    'Saving_Accounts', 'Foreign_Currency_Account', 'Business_Lending'
]
correlation_matrix = df[numerical_cols].corr()

mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

plt.figure(figsize=(9, 7))
heatmap = sns.heatmap(
    correlation_matrix,
    mask=mask,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",  # or "vlag", "Spectral", "RdBu"
    square=True,
    linewidths=0.5,
    cbar_kws={"shrink": 0.8, "aspect": 30}
)

plt.title("Correlation Matrix of Financial Metrics", fontsize=18, fontweight='bold', color='navy', pad=20)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
